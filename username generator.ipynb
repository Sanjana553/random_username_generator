```python
import random
import string
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to generate a random username
def generate_username(include_numbers=True, include_special_chars=True, length=8):
    adjectives = ["Cool", "Happy", "Brave", "Clever", "Swift", "Loyal", "Jolly", "Fierce"]
    nouns = ["Tiger", "Dragon", "Eagle", "Wolf", "Panther", "Falcon", "Lion", "Shark"]

    adj = random.choice(adjectives)
    noun = random.choice(nouns)
    username = adj + noun

    if include_numbers:
        username += str(random.randint(10, 99))

    if include_special_chars:
        username += random.choice(string.punctuation)

    return username[:length]

# Function to generate and display usernames
def on_generate_button_click(b):
    clear_output(wait=True)
    display(ui)  # Redisplay UI after clearing

    num_usernames = num_usernames_slider.value
    length = length_slider.value
    include_numbers = numbers_checkbox.value
    include_special_chars = special_chars_checkbox.value

    usernames = [generate_username(include_numbers, include_special_chars, length) for _ in range(num_usernames)]

    print("\n🎉 Generated Usernames 🎉")
    for uname in usernames:
        print(f"🔹 {uname}")

# Creating interactive widgets
numbers_checkbox = widgets.Checkbox(value=True, description="Include Numbers")
special_chars_checkbox = widgets.Checkbox(value=True, description="Include Special Characters")
length_slider = widgets.IntSlider(value=8, min=5, max=15, step=1, description="Length")
num_usernames_slider = widgets.IntSlider(value=5, min=1, max=20, step=1, description="Number of Usernames")
generate_button = widgets.Button(description="Generate Usernames", button_style="success")

# Attach event handler to button
generate_button.on_click(on_generate_button_click)

# Arrange UI elements
ui = widgets.VBox([
    numbers_checkbox,
    special_chars_checkbox,
    length_slider,
    num_usernames_slider,
    generate_button
])

# Display UI
display(ui)
```
